```python
from pytrends.request import TrendReq
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import folium
```

```python
keywords = ['던파', '던전앤파이터', '던파 M', '던파모바일', '던파 PC']
pytrends = TrendReq(hl = 'ko', tz = 360)
pytrends.build_payload(keywords, cat = 0, timeframe = '2022-01-01 2022-04-24', geo = 'KR', gprop = 'youtube')
getData = pytrends.interest_over_time()
print(getData)
```

```python
getData['total'] = getData.sum(axis = 'columns')
```

```python
getData
```

```python
getData['total'].plot()
```

```python
xs = getData.index.to_list()#dy_day(데이터 프레임)의 index(날짜, 시간)를 리스트로 저장
ys = getData['total'].to_list() #dy_day(테이터 프레임)의 volume 필드를 리스트로 저장
plt.figure(figsize=(10, 6)) #그래프 크기 지정
plt.xlabel('Date') #그래프 x축 이름(label) 지정
plt.ylabel('Total') #그래프 y축 이름(label) 지정
plt.bar(xs, ys, width=0.6) #리스트로 저장한 xs와 ys로 막대(bar) 그래프 플롯
```
